```python
# IMPORTING THE DATA:
import pandas as pd  # library for data handling, usefull data stuctures here
player_data = pd.read_csv('data.csv')  # imports the CSV file with the player data from FIFA 19
player_data.set_index('Name', inplace=True)  # sets players's name as index
player_data.drop(['Unnamed: 0', 'ID'], axis=1, inplace=True)  # remove useless columns

print(player_data.shape)  # data dimensions
print(player_data.columns) # columns available
```

```python
display(player_data)
```

## Data Preparation (the most easy you'll see)

We are going to filter the players for our analysis, since the goal is to show the algorithm working we'll only keep the best players, because most of us are familiar with theirs characteristics.

We'll also remove Goal Keepers because no one cares about them. JK, just because they are a very specific group that would probably mess up the analysis.

We'll only focus on a part of the features, as you can see bellow.

```python
# We are going to filter the players for our analysis, the c
player_data_raw = player_data  # we'll keep the raw data for later
player_data = player_data[player_data['Overall'] > 82]  # filter the best players
player_data = player_data[player_data['Position'] != 'GK'] # removes Goal Keepers
features_to_keep = ['Position', 'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
                    'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength',
                    'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
                    'Marking', 'StandingTackle', 'SlidingTackle']
player_data = player_data[features_to_keep]  # keep only the chosen features
print(player_data.shape)  # data dimensions
data_matrix = player_data[list(player_data.columns)[2:]].values  #  gets the feature's data in a matrix format for model input
```

```python
display(player_data)
display(data_matrix)
```

## Data Modeling

Now we'll start to get into the Machine Learning models...

We'll apply a k-means algo to cluster the players, based on theirs features. We'll use a python library with these models implemented, scikit-learn.

The number of cluster will be 5, this choice was made by me, knowing how do generally players group according to their characteristics. In a real scenario, if we don't have a clear insight of the problem several number of clusters should be tested to check how to best fit the data. This topic has enough content for a talk itself.

```python
from sklearn import cluster  # import the library to use
model = cluster.KMeans(n_clusters=5)  # initialize the model class with 5 clusters as parameter
model.fit(data_matrix)  # train the model with our pre-prepared data
labels = model.labels_  # check which cluster was assigned to the players
results = pd.DataFrame(labels, index=player_data.index, columns=['cluster_number'])  # save the results in a pandas dataframe
display(results.head(10))  # display the first 10 players
```

```python
player_data = player_data.join(results)  # join with the rest of the data

for nr in range(5):
    print(f'\nPlayers in cluster number {nr}')
    print(player_data[player_data['cluster_number'] == nr].index)
```

Let's check what happens when we show to our model players not present in the training data...

```python
players_to_predict = player_data_raw.loc[['Óliver Torres', 'J. Mathieu', 'Rafa']]  # extracting 3players from the raw data
players_to_predict = players_to_predict[features_to_keep]  #  keep only the features known by the model
display(players_to_predict)
```

```python
prediction_data_matrix = players_to_predict[list(players_to_predict.columns)[2:]].values  # prepare the data format to input in the mdel
predicted_labels = model.predict(prediction_data_matrix)  # make the predictions
predictions = pd.DataFrame(predicted_labels, index=players_to_predict.index, columns=['cluster_number'])  # save the results in a pandas dataframe
display(predictions)
```
